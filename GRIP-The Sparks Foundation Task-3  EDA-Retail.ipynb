```python
plt.figure(figsize=(8,5))
sns.kdeplot(SampleSuperstore['Sales'],color='Blue', label='Sales',shade=True,bw=25)
sns.kdeplot(SampleSuperstore['Profit'],color='red', label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

```python
off_subcat = SampleSuperstore[SampleSuperstore['Category'] == 'Office Supplies']
fur_subcat = SampleSuperstore[SampleSuperstore['Category'] == 'Furniture']
tech_subcat = SampleSuperstore[SampleSuperstore['Category'] == 'Technology']

cat_profit_df= pd.DataFrame( [off_subcat['Profit'].mean(), fur_subcat['Profit'].mean(), tech_subcat['Profit'].mean()],
                          index=['Office Supplies','Furniture', 'Technology'])

cat_profit_df.plot(kind='pie', subplots=True, figsize=(16,8), autopct='%1.1f%%',
                   title='Categories Profit Comparision')
```

```python
SampleSuperstore_seg = SampleSuperstore.groupby(['Segment'])[['Sales', 'Profit', 'Discount']].mean()
SampleSuperstore_seg
```

```python
SampleSuperstore_seg.plot.pie(subplots = True,
                             autopct = '%1.1f%%',
                             figsize = (18,20),
                             startangle = 90,
                             shadow = True,
                             labels = SampleSuperstore_seg.index)
plt.title('Segmentwise analysis')
```

```python
con_seg = SampleSuperstore[SampleSuperstore['Segment'] == 'Consumer']
cor_seg = SampleSuperstore[SampleSuperstore['Segment'] == 'Corporate']
home_seg = SampleSuperstore[SampleSuperstore['Segment'] == 'Home Office']


fig = plt.figure(figsize=(20,5))

plt.subplot(131)
plt.scatter(con_seg["Sales"] , con_seg["Profit"])
plt.title('Consumer Sales vs Profit')

plt.subplot(132)
plt.scatter(cor_seg["Sales"] , cor_seg["Profit"])
plt.title('Corporate Sales vs Profit')

plt.subplot(133)
plt.scatter(home_seg["Sales"] , home_seg["Profit"])
plt.title('Home Office Sales vs Profit')

plt.show()
```
